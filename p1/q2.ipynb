```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movieReplicationSet.csv')
```

```python
df
```

```python
df.columns
```

```python
# 从列名提取年份
years = [int(re.search(r'\((\d{4})\)', col).group(1)) for col in df.iloc[:,0:400].columns]

# 计算中位数
median_year = np.median(years)
print(median_year)
# 根据中位数进行分类
older_movies = [col for col, year in zip(df.columns, years) if year <= median_year]
newer_movies = [col for col, year in zip(df.columns, years) if year > median_year]
```

```python
old_movies = df[older_movies]
new_movies = df[newer_movies]
```

```python
flattened_old = [item for sublist in old_movies.values.T.tolist() for item in sublist]
flattened_new = [item for sublist in new_movies.values.T.tolist() for item in sublist]

```

```python
t_stat, p_value = ttest_ind(flattened_new, flattened_old, nan_policy='omit')
print(t_stat, p_value)
```

```python
t_stat, p_value = ttest_ind(flattened_new, flattened_old, nan_policy='omit',alternative='greater')
print(t_stat, p_value)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=[flattened_new, flattened_old])
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.title('Boxplot of Group 1 and Group 2')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=[flattened_new, flattened_old])
plt.xticks([0, 1], ['newer movies', 'old movies'])
plt.title('Rating of newer and older movies')
plt.show()
```

```python
import numpy as np
from scipy.stats import levene

# 生成两组模拟数据


# 执行 Levene's test
stat, p = levene(pd.Series(flattened_new).dropna(), pd.Series(flattened_old).dropna())

print(f"Levene's Test Statistic: {stat}")
print(f"P-Value: {p}")

if p < 0.005:
    print("Variances are significantly different")
else:
    print("Variances are not significantly different")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movieReplicationSet.csv')
df
```

```python
df.columns[475]
# 'Are you an only child? (1: Yes; 0: No; -1: Did not respond)'
```

```python
only_child_col = df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)']
```

```python
only_child = only_child_col[only_child_col == 1].index
siblings = only_child_col[only_child_col == 0].index
```

```python
only_child_movies = df.iloc[only_child,0:400]
siblings_movies = df.iloc[siblings,0:400]
```

```python
count = 0
for i in only_child_movies.columns:
    t_stat, p_value = ttest_ind(only_child_movies[i], siblings_movies[i], nan_policy='omit')
    if p_value <= 0.005:
        count += 1
print(count/400)
print(count)
```

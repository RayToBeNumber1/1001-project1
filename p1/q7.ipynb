```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movieReplicationSet.csv')
```

```python
df.columns[476]
```

```python
alone = df[df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 1].index
social = df[df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'] == 0].index
```

```python
alone_movies = df['The Wolf of Wall Street (2013)'].iloc[alone]
social_movies = df['The Wolf of Wall Street (2013)'].iloc[social]
```

```python
t_stat, p_value = ttest_ind(social_movies, alone_movies, nan_policy='omit', alternative='less')
```

```python
p_value
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=[social_movies, alone_movies])
plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.title('Violin plot of Group 1 and Group 2')
plt.show()
```

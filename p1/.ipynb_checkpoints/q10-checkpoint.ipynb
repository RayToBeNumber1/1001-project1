```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp
from scipy.stats import f_oneway
```

```python
df = pd.read_csv('movieReplicationSet.csv')
```

```python
movie_ls = ['Star Wars', 'Harry Potter', 'The Matrix', 'Indiana Jones', 
            'Jurassic Park', 'Pirates of the Caribbean', 'Toy Story', 'Batman']
```

```python
dict_movies = {}
for i in movie_ls:
    dict_movies[i] = []
    
print(dict_movies)

    
```

```python
for col in df.columns:
    for movie in movie_ls:
        if movie in col:
            dict_movies[movie].append(df[col].dropna())
            break

        
```

```python
len(dict_movies['Star Wars'])
```

```python
count = 0
for movie in dict_movies:
    f_statistic, p_value  = f_oneway(*dict_movies[movie])
    if p_value <= 0.005:
        count += 1
        print(movie)
print(count)
        
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=[*dict_movies['Harry Potter']])
# plt.xticks([0, 1], ['Group 1', 'Group 2'])
plt.title('Violin plot of Harry Potter (SAME)')
plt.show()
```
